```python
import jieba
import re
import json
```

```python
# text = "今天天气很好，我们一起去公园玩吧！"
with open("data.txt") as f:
    text = f.read()
# 精确模式分词
def remove_punctuation(text):
    text = re.sub(r'[^\w\s]', '', text)  # 去除标点符号
    return text
text = remove_punctuation(text)
words = jieba.lcut(text)
words = [w for w in words if w != '\n' and w != ' ' and w != '\u3000']
print(len(words))
```

```python
processed_text = " ".join(words)
with open('processed_data.txt','w') as f:
    f.write(processed_text)
```

```python
from collections import Counter

# 分词后的结果
# 统计词频
word_freq = Counter(words)
print(word_freq.most_common(10))  # 输出出现次数最多的前10个词
```

```python
word_freq_dict = dict(word_freq)
with open("words_freq.json",'w',) as f:
    json.dump(word_freq_dict,f,ensure_ascii=False)
```

```python
with open("words_freq.json",'r') as f:
    words_freq = json.load(f)
word2id = {key:i for i,key in enumerate(words_freq)}
with open("word2id.json","w") as f:
    json.dump(word2id,f,ensure_ascii=False)
```
