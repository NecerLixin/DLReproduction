```python
import pandas as pd
import json
import numpy as np
```

```python
train_data = pd.read_csv("FB15K/train.txt", delimiter="\t", names=["h", "l", "t"])
test_data = pd.read_csv("FB15K/test.txt", delimiter="\t", names=["h", "l", "t"])
valid_data = pd.read_csv("FB15K/valid.txt", delimiter="\t", names=["h", "l", "t"])
```

```python
all_data = pd.concat([train_data, test_data, valid_data], axis=0)
all_triples = all_data.drop_duplicates().values.tolist()
```

```python
all_triples = [tuple(s) for s in all_triples]
all_triples = set(all_triples)
all_triples
```

```python
# 获取所有的实体，entity2index
all_entities = set(all_data["h"].unique().tolist() + all_data["t"].unique().tolist())
entity2index = {entity: i for i, entity in enumerate(all_entities)}
# 获取所有的关系，label2index
all_labels = set(all_data["l"].tolist())
label2index = {label: i for i, label in all_labels}
with open("FB15k/entity2index.json", "w") as f:
    json.dump(entity2index, f, ensure_ascii=False)
with open("FB15k/label2index.json", "w") as f:
    json.dump(label2index, f, ensure_ascii=False)
```
